# âœï¸ Escribir informaciÃ³n en un archivo

Cuando querÃ©s **guardar informaciÃ³n para usar mÃ¡s adelante**, podÃ©s escribirla en un archivo de texto.

En este ejemplo vas a guardar una lista de tareas en un archivo llamado `tareas.txt`.


```python
# Lista de tareas
tareas = ["barrer", "lavar los platos", "hacer la cama"]

# Abrimos el archivo en modo escritura
with open("tareas.txt", "w") as archivo:
    for tarea in tareas:
        archivo.write(tarea + "\n")

print("Tareas guardadas en el archivo.")
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `with open("tareas.txt", "w")` abre (y si no existe, crea) un archivo de texto para escritura.

ğŸ“Œ `.write()` escribe cada lÃ­nea. Agregamos `\n` para que cada tarea vaya en una lÃ­nea nueva.

ğŸ“Œ El bloque `with` se encarga de **cerrar el archivo automÃ¡ticamente**.

ğŸ¯ PodÃ©s ver en la secciÃ³n "Archivos" de Colab el archivo `tareas.txt` que se ha creado. Si le das doble click, verÃ¡s su contenido en un nuevo panel de Colab.


# ğŸ“– Leer informaciÃ³n desde un archivo

Si guardaste datos en un archivo, luego podÃ©s **leerlos y usarlos en tu programa**.

En este ejemplo vamos a leer las tareas que se guardaron en `tareas.txt`.


```python
# Abrimos el archivo en modo lectura
with open("tareas.txt", "r") as archivo:
    print("ğŸ“„ Tareas registradas:")
    for linea in archivo:
        print("-", linea.strip())
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `open("tareas.txt", "r")` abre el archivo en **modo lectura**.

ğŸ“Œ El bucle `for` recorre cada lÃ­nea del archivo.

ğŸ“Œ `.strip()` elimina el salto de lÃ­nea (`\n`) que se guardÃ³ al escribir.

# ğŸ›¡ï¸ Manejar errores con try-except

Cuando algo puede fallar (por ejemplo, una divisiÃ³n, una conversiÃ³n o un acceso a archivo),

En este ejemplo, intentamos convertir una entrada en nÃºmero.


```python
entrada = input("IngresÃ¡ un nÃºmero: ")

try:
    numero = int(entrada)
    print("El doble es:", numero * 2)
except:
    print("Eso no era un nÃºmero vÃ¡lido.")
```

# âœ… SoluciÃ³n - Registro validado de notas en archivo

```python
# Bucle de ingreso
while True:
    entrada = input("IngresÃ¡ una nota del 1 al 10 (o 'salir'): ").strip()

    if entrada.lower() == "salir":
        break

    try:
        nota = int(entrada)
        if 1 <= nota <= 10:
            with open("notas.txt", "a") as archivo:
                archivo.write(str(nota) + "\n")
            print("Nota guardada.")
        else:
            print("La nota debe estar entre 1 y 10.")
    except:
        print("IngresÃ¡ un nÃºmero vÃ¡lido o escribÃ­ 'salir'.")
```

# âœ… SoluciÃ³n - Mostrar todas las notas cargadas

```python
print("\nNotas registradas:")

try:
    with open("notas.txt", "r") as archivo:
        for linea in archivo:
            print("-", linea.strip())
except:
    print("No se encontrÃ³ el archivo 'notas.txt'.")
```

ğŸ“Œ El programa repite el ingreso hasta que el usuario escribe "salir".
