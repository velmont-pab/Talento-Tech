# ğŸ“‹ Recorrer una lista con for

Hasta ahora recorrÃ­as listas usando `while` y un contador manual.  
Ahora vas a ver cÃ³mo el bucle `for` te permite **hacer lo mismo de forma mÃ¡s clara y automÃ¡tica**.

En este ejemplo vamos a recorrer una lista de nombres e imprimir cada uno en pantalla.


```python
# Lista de nombres
nombres = ["AilÃ©n", "Bautista", "Camila", "DamiÃ¡n"]

# Recorrido con for
for nombre in nombres:
    print(f"Hola, {nombre}!")
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ La variable `nombre` **toma uno por uno** los valores de la lista `nombres`.

# ğŸ“‹ Recorrer una lista con for

Ya sabÃ©s que con `while` podÃ©s recorrer listas usando un Ã­ndice.  
Ahora vas a ver cÃ³mo el mismo resultado se puede lograr **de forma mÃ¡s sencilla** usando `for`.

En este ejemplo vamos a mostrar los nombres de una lista con ambas estructuras.


```python
# Lista de nombres
nombres = ["AilÃ©n", "Bautista", "Camila", "DamiÃ¡n"]

# Recorrido con while
print("Recorrido con while:")
i = 0
while i < len(nombres):
    print(nombres[i])
    i += 1

# Recorrido con for
print("\nRecorrido con for:")
for nombre in nombres:
    print(nombre)
```

# ğŸ” Â¿QuÃ© diferencias vemos?

ğŸ“Œ Ambos bucles recorren la lista completa y muestran los mismos valores.

ğŸ“Œ Con `while` necesitÃ¡s:
- Declarar una variable Ã­ndice (`i = 0`)
- Acceder manualmente a cada elemento (`nombres[i]`)
- Acordarte de aumentar el Ã­ndice (`i += 1`)

ğŸ“Œ Con `for`, el recorrido es **automÃ¡tico y mÃ¡s legible**.

ğŸ’¡ Siempre que trabajes con listas (o textos), `for` suele ser la opciÃ³n mÃ¡s clara.

# ğŸ“‹ Recorrer un texto letra por letra

A veces no queremos recorrer toda una lista, sino **detenernos cuando ocurre algo**.

El comando `break` interrumpe el bucle en el momento en que se cumple una condiciÃ³n.

En este ejemplo vamos a detenernos cuando encontremos el nombre "Camila".


```python
# Lista de nombres
nombres = ["AilÃ©n", "Bautista", "Camila", "DamiÃ¡n", "Eugenia"]

# Recorremos hasta encontrar "Camila"
for nombre in nombres:
    print(f"Revisando: {nombre}")
    if nombre == "Camila":
        print("Â¡Encontramos a Camila!")
        break
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ El bucle recorre los nombres uno por uno.

ğŸ“Œ Cuando encuentra `"Camila"`, se muestra el mensaje y el bucle **termina inmediatamente**.

ğŸ“Œ `break` se usa cuando querÃ©s **salir de un bucle antes de tiempo**, por ejemplo:
- Al encontrar un error
- Al hallar un valor buscado
- Al validar condiciones especiales

ğŸ¯ ProbÃ¡ cambiar el nombre de corte por otro que estÃ© al final o que no estÃ© en la lista.

# ğŸ“‹ Usar `continue` para saltear valores no vÃ¡lidos

En algunas listas puede haber valores incompletos, vacÃ­os o incorrectos.

El comando `continue` permite **saltar esa iteraciÃ³n** y pasar directamente al siguiente valor, sin detener el bucle.

En este ejemplo vamos a ignorar los nombres vacÃ­os o que contienen solo espacios.


```python
# Lista con algunos valores vacÃ­os
nombres = ["AilÃ©n", "", "  ", "Camila", "DamiÃ¡n"]

# Recorremos y mostramos solo los nombres vÃ¡lidos
for nombre in nombres:
    if nombre.strip() == "":
        continue  # Salteamos los nombres vacÃ­os
    print(f"Nombre vÃ¡lido: {nombre.strip().title()}")
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `nombre.strip()` elimina los espacios al principio y al final.

ğŸ“Œ Si el nombre queda vacÃ­o despuÃ©s de aplicar `.strip()`, el bucle salta esa iteraciÃ³n con `continue`.

ğŸ“Œ Solo se muestran los nombres que **contienen texto vÃ¡lido**.

ğŸ“Œ Este patrÃ³n es ideal para **limpiar listas antes de procesarlas**, sin interrumpir el flujo del programa.

ğŸ¯ PodÃ©s agregar mÃ¡s valores a la lista, incluso con errores o letras mezcladas, y ver cÃ³mo se comporta el filtro.
