#  Usar el m贸dulo random para lanzar un dado

El m贸dulo `random` permite generar **valores aleatorios**.

En este ejemplo lo usamos para simular el lanzamiento de un dado de 6 caras.


````python
# Importamos el m贸dulo
import random

# Lanzamos el dado
dado = random.randint(1, 6)

print("El dado cay贸 en:", dado)
````

#  Usar el m贸dulo math para c谩lculos matem谩ticos

El m贸dulo `math` ofrece funciones matem谩ticas 煤tiles, como **ra铆z cuadrada**, **redondeo** y constantes como `pi`.

En este ejemplo vamos a calcular la ra铆z cuadrada y redondear resultados.


````python
# Importamos el m贸dulo
import math

# N煤mero de ejemplo
numero = 20

# Ra铆z cuadrada
raiz = math.sqrt(numero)

# Redondeamos hacia arriba y hacia abajo
arriba = math.ceil(raiz)
abajo = math.floor(raiz)

print(f"Ra铆z cuadrada de {numero}: {raiz}")
print(f"Redondeo hacia arriba: {arriba}")
print(f"Redondeo hacia abajo: {abajo}")
````

#  驴Qu茅 hace este c贸digo?

  `math.ceil(n)` redondea **hacia arriba** (al entero mayor).

  `math.floor(n)` redondea **hacia abajo** (al entero menor).

 Prob谩 con distintos n煤meros, incluyendo decimales o negativos (驴qu茅 pasa con `sqrt(-4)`?).

#  Usar el m贸dulo datetime para mostrar fecha y hora

El m贸dulo `datetime` permite trabajar con fechas y horas.

En este ejemplo vamos a mostrar **la fecha y hora actual del sistema**.


````python
# Importamos solo la clase datetime del m贸dulo
from datetime import datetime

# Obtenemos la fecha y hora actual
ahora = datetime.now()

# Mostramos fecha y hora con formato
print("Fecha y hora actual:")
print(ahora.strftime("%d/%m/%Y %H:%M:%S"))
````

#  驴Qu茅 hace este c贸digo?

 `datetime.now()` obtiene la fecha y hora actuales.

  `.strftime(...)` permite **formatear** la salida como texto.

*  `%d` = d铆a / `%m` = mes / `%Y` = a帽o  
*   `%H` = hora / `%M` = minuto / `%S` = segundo

 Prob谩 cambiar el formato o mostrar solo la hora, o solo la fecha.
