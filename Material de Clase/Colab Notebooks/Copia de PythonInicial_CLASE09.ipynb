# 游깴 Variables locales y globales

Las variables pueden existir **dentro o fuera** de una funci칩n.

Las que se crean dentro son **locales** (viven solo ah칤).  
Las que est치n afuera son **globales** (pueden usarse en todo el programa, con cuidado).


```python
# Variable global
mensaje = "Hola desde afuera"

def mostrar_mensaje():
    mensaje = "Hola desde adentro"
    print(mensaje)

mostrar_mensaje()

print(mensaje)  # La global no cambi칩
```

# 游댌 쯈u칠 hace este c칩digo?

游늷 Dentro de la funci칩n se crea una variable `mensaje` local.

游늷 Aunque tenga el mismo nombre, **no modifica** la variable global.

游늷 Las variables locales **solo existen dentro de la funci칩n**.  
Fuera de ella, no tienen efecto.

# 游댃 Mutabilidad: modificar una lista desde una funci칩n

Cuando pas치s una lista (o un diccionario) a una funci칩n, est치s pasando **una referencia al mismo objeto**.

Por eso, si la funci칩n lo modifica, el cambio **afecta a la variable original**.


```python
# Funci칩n que modifica una lista
def agregar_producto(lista, producto):
    lista.append(producto)
    print(f"Producto '{producto}' agregado.")

# Lista de productos
productos = ["leche", "pan"]

# Llamamos a la funci칩n
agregar_producto(productos, "arroz")

print("\nLista despu칠s de llamar a la funci칩n:")
print(productos)
```

# 游댌 쯈u칠 hace este c칩digo?

游늷 La lista `productos` fue modificada **dentro de la funci칩n**, y el cambio qued칩 reflejado afuera.

游늷 Esto ocurre porque las listas son **mutables**: pueden cambiarse sin crear una nueva copia.

游늷 Si no quer칠s que la funci칩n modifique la lista original, pod칠s pasarle una **copia**: `lista.copy()`

Esta es una posible soluci칩n al desaf칤o propuesto:

```python
# Funci칩n para agregar un producto
def agregar(productos):
    nuevo = input("Nombre del producto a agregar: ").strip()
    if nuevo == "":
        print("No se puede agregar un nombre vac칤o.")
    else:
        productos.append(nuevo)
        print(f"Producto '{nuevo}' agregado.")

# Funci칩n para mostrar productos
def mostrar(productos):
    if not productos:
        print("La lista est치 vac칤a.")
    else:
        print("Lista de productos:")
        for producto in productos:
            print("-", producto)

# Funci칩n para eliminar un producto
def eliminar(productos):
    a_eliminar = input("Nombre del producto a eliminar: ").strip()
    if a_eliminar in productos:
        productos.remove(a_eliminar)
        print(f"Producto '{a_eliminar}' eliminado.")
    else:
        print("Ese producto no est치 en la lista.")

# Programa principal
def menu():
    productos = []

    while True:
        print("\n--- Men칰 ---")
        print("1. Agregar producto")
        print("2. Mostrar productos")
        print("3. Eliminar producto")
        print("4. Salir")

        opcion = input("Eleg칤 una opci칩n: ").strip()

        if opcion == "1":
            agregar(productos)
        elif opcion == "2":
            mostrar(productos)
        elif opcion == "3":
            eliminar(productos)
        elif opcion == "4":
            print("춰Hasta la pr칩xima!")
            break
        else:
            print("Opci칩n inv치lida.")

# Ejecutamos el programa
menu()
```

# 游댌 쯈u칠 hace este c칩digo?

游늷 Cada acci칩n est치 separada en una funci칩n: esto **ordena el c칩digo** y lo hace m치s f치cil de mantener.

游늷 La lista `productos` se pasa como par치metro para que todas las funciones trabajen sobre el mismo conjunto de datos.

游늷 Se validan entradas vac칤as y productos inexistentes al eliminar.

游늷 El bucle principal repite el men칰 hasta que el usuario elige salir.
