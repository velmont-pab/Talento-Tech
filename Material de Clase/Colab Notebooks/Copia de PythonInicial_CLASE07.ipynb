# ğŸ› ï¸ Agregar, quitar y ordenar elementos de una lista

En este bloque vas a ver cÃ³mo usar algunos mÃ©todos importantes de las listas en acciÃ³n:

- `.append()` para agregar al final
- `.remove()` para eliminar un valor
- `.sort()` para ordenar la lista

Vamos a trabajar con una lista de productos que se van cargando, eliminando y luego ordenando alfabÃ©ticamente.


```python
# Lista vacÃ­a
productos = []

# Agregamos algunos productos
productos.append("arroz")
productos.append("fideos")
productos.append("harina")
productos.append("leche")

print("Lista despuÃ©s de agregar:")
print(productos)

# Quitamos un producto
productos.remove("fideos")

print("\nLista despuÃ©s de quitar 'fideos':")
print(productos)

# Ordenamos alfabÃ©ticamente
productos.sort()

print("\nLista ordenada alfabÃ©ticamente:")
print(productos)
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `append()` agrega elementos al final de la lista, en el orden que llegan.

ğŸ“Œ `remove("fideos")` elimina el valor si estÃ¡ presente (si no, darÃ­a error).

ğŸ“Œ `sort()` ordena los elementos en orden alfabÃ©tico (de menor a mayor si fueran nÃºmeros).

# ğŸ› ï¸ Quitar por posiciÃ³n, insertar e incorporar varias tareas

Vamos a usar tres mÃ©todos diferentes para modificar una lista de tareas:

- `.pop()` â†’ elimina un elemento **por posiciÃ³n**
- `.insert()` â†’ agrega un elemento **en una posiciÃ³n especÃ­fica**
- `.extend()` â†’ agrega varios elementos juntos, desde otra lista


```python
# Lista inicial de tareas
tareas = ["barrer", "lavar los platos", "hacer la cama"]

print("Lista original:")
print(tareas)

# Quitamos la segunda tarea (posiciÃ³n 1)
tareas.pop(1)

print("\nDespuÃ©s de eliminar la tarea en posiciÃ³n 1:")
print(tareas)

# Insertamos una nueva tarea en primer lugar
tareas.insert(0, "sacar la basura")

print("\nDespuÃ©s de insertar una tarea al principio:")
print(tareas)

# Extendemos la lista con otras tareas
tareas_extra = ["regar las plantas", "ordenar el escritorio"]
tareas.extend(tareas_extra)

print("\nDespuÃ©s de extender con mÃ¡s tareas:")
print(tareas)
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `pop(1)` elimina el segundo elemento (recordÃ¡ que la cuenta empieza en 0).

ğŸ“Œ `insert(0, ...)` pone una nueva tarea en el primer lugar.

ğŸ“Œ `extend(lista)` agrega todos los elementos de otra lista al final, sin anidar listas.

ğŸ¯ ProbÃ¡ cambiar el nÃºmero en `pop()`, o usar `insert()` en otra posiciÃ³n para ver cÃ³mo se reordenan las tareas.


# ğŸ› ï¸ Buscar, contar y limpiar una lista

En el ejemplo siguiente vas a ver cÃ³mo:

- `index()` te dice **en quÃ© posiciÃ³n estÃ¡ un elemento**
- `count()` te dice **cuÃ¡ntas veces aparece**
- `clear()` **borra todos los elementos** de la lista


```python
# Lista con invitados
invitados = ["Camila", "Juan", "SofÃ­a", "Camila", "DamiÃ¡n"]

print("Lista de invitados:")
print(invitados)

# Buscamos la primera posiciÃ³n donde aparece "Camila"
posicion = invitados.index("Camila")
print(f"\nCamila aparece por primera vez en la posiciÃ³n {posicion}.")

# Contamos cuÃ¡ntas veces aparece "Camila"
cantidad = invitados.count("Camila")
print(f"Camila aparece {cantidad} veces en la lista.")

# Limpiamos la lista de invitados
invitados.clear()
print("\nDespuÃ©s de usar .clear():")
print(invitados)
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `index("Camila")` devuelve la **primera posiciÃ³n** en la que aparece.

ğŸ“Œ `count("Camila")` cuenta **cuÃ¡ntas veces** aparece ese nombre.

ğŸ“Œ `clear()` deja la lista completamente vacÃ­a: `[]`.

# ğŸ› ï¸ Tuplas: datos que no se modifican

Las **tuplas** se parecen a las listas, pero tienen una diferencia clave: **no se pueden modificar**.

Se usan cuando querÃ©s guardar datos que **no deberÃ­an cambiar**.

En este bloque vas a ver cÃ³mo crear una tupla y recorrerla con un `for`.


```python
# Creamos una tupla con dÃ­as de la semana
dias = ("lunes", "martes", "miÃ©rcoles", "jueves", "viernes")

# Mostramos cada dÃ­a
for dia in dias:
    print(f"Hoy es {dia}")
```

# ğŸ” Â¿QuÃ© diferencia tiene con una lista?

âœ”ï¸ A simple vista, funciona igual: podÃ©s recorrerla con `for` y acceder por posiciÃ³n.

âŒ Pero si intentÃ¡s modificarla (por ejemplo, `dias[0] = "Domingo"`), vas a obtener un **error**.

âœ”ï¸ Las tuplas se definen con **parÃ©ntesis** `()`  
Las listas se definen con **corchetes** `[]`

ğŸ¯ UsÃ¡ tuplas cuando querÃ©s **proteger datos** que no deberÃ­an cambiar, como dÃ­as de la semana, meses, o claves predefinidas.


# ğŸ› ï¸ Trabajar con partes de una lista (slicing)

A veces no necesitÃ¡s toda una lista, sino **una parte**: los primeros elementos, los Ãºltimos, o un grupo intermedio.

PodÃ©s obtener sublistas usando la notaciÃ³n `[inicio:fin]`  
(Â¡el elemento de la posiciÃ³n final no se incluye!)


```python
# Lista con notas de un estudiante
notas = [7, 8, 10, 6, 9, 5, 8]

print("Todas las notas:")
print(notas)

# Primeras tres notas
print("\nPrimeras tres:")
print(notas[:3])

# Ãšltimas dos notas
print("\nÃšltimas dos:")
print(notas[-2:])

# Notas del medio
print("\nNotas de la posiciÃ³n 2 a la 4:")
print(notas[2:5])
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `[0:3]` devuelve los elementos en posiciones 0, 1 y 2.

ğŸ“Œ Si no indicÃ¡s el inicio (`[:n]`), empieza desde el principio.  
Si no indicÃ¡s el final (`[n:]`), va hasta el final.

ğŸ“Œ TambiÃ©n podÃ©s usar Ã­ndices negativos para contar desde atrÃ¡s (`[-2:]`).

ğŸ¯ ProbÃ¡ cambiar los rangos y analizÃ¡ quÃ© sublista se genera.


# DesafÃ­o - Limpiar y mostrar una lista de nombres

Queremos registrar el nombre de las personas que se anotan a un taller.  Pero a veces escriben mal o dejan el campo vacÃ­o.

Tu desafÃ­o es:

1. Usar un bucle para **cargar nombres** (por ejemplo, 5)
2. Ignorar los vacÃ­os o espacios en blanco
3. Guardar los nombres vÃ¡lidos usando `.append()`
4. Ordenar la lista alfabÃ©ticamente con `.sort()`
5. Mostrar solo los **primeros 3 nombres vÃ¡lidos**, usando slicing

ğŸ¯ Bonus: mostrar la cantidad total de nombres registrados correctamente.

(Intenta resolverlo antes de mirar la posible soluciÃ³n que aparece mÃ¡s abajo!)

```python
# Escribe aqui tu programa
```

----
Esta es una posible soluciÃ³n al desafÃ­o propuesto:

```python
# Inicializamos la lista y el contador
nombres = []
cantidad_deseada = 5
cargados = 0

# Bucle para cargar nombres vÃ¡lidos
while cargados < cantidad_deseada:
    nombre = input(f"IngresÃ¡ el nombre #{cargados + 1}: ").strip()

    if nombre == "":
        print("No se aceptan nombres vacÃ­os.")
        continue

    nombres.append(nombre.title())  # Formateamos y agregamos
    cargados += 1

# Ordenamos la lista
nombres.sort()

# Mostramos resultados
print("\nLista ordenada de nombres vÃ¡lidos:")
print(nombres)

# Mostramos solo los primeros 3 con slicing
print("\nPrimeros 3 nombres en orden alfabÃ©tico:")
print(nombres[:3])

# Total cargado
print(f"\nTotal de nombres registrados: {len(nombres)}")
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ Se usÃ³ `.strip()` para eliminar espacios y evitar entradas vacÃ­as.

ğŸ“Œ Se formateÃ³ cada nombre con `.title()` antes de agregarlo.

ğŸ“Œ El contador `cargados` asegura que se registren exactamente 5 nombres vÃ¡lidos.

ğŸ“Œ `.sort()` organiza los nombres alfabÃ©ticamente.

ğŸ“Œ `nombres[:3]` devuelve los primeros 3 sin necesidad de otro bucle.
