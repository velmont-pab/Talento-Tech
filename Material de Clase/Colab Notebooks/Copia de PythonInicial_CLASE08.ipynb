# ğŸ“‡ Crear y usar un diccionario

Un diccionario te permite guardar informaciÃ³n con una estructura clara: **clave: valor**

En este ejemplo vas a crear un diccionario con los datos de una persona y acceder a cada valor usando su clave.


```python
# Creamos un diccionario con algunos datos
persona = {
    "nombre": "MarÃ­a",
    "edad": 30,
    "ciudad": "Rosario"
}

# Mostramos los valores accediendo por clave
print("Nombre:", persona["nombre"])
print("Edad:", persona["edad"])
print("Ciudad:", persona["ciudad"])

# Agregamos una nueva clave:valor
persona["email"] = "maria@gmail.com"

print("\nDespuÃ©s de agregar el email:")
print(persona)
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ Cada entrada tiene una **clave Ãºnica** y un valor asociado.

ğŸ“Œ Para acceder a un valor, usÃ¡s `diccionario["clave"]`

ğŸ“Œ PodÃ©s **agregar un nuevo dato** simplemente escribiendo una nueva clave.

ğŸ¯ ProbÃ¡ cambiar los valores o agregar otras claves como "telÃ©fono" o "profesiÃ³n".


# ğŸ“‡ Usar `.get()` y `.pop()` en un diccionario

`.get()` y `.pop()` son dos mÃ©todos muy Ãºtiles de los diccionarios:

- `.get(clave)` devuelve el valor si existe, o `None` si no estÃ¡ (sin generar error)
- `.pop(clave)` elimina un par clave:valor y devuelve el valor que tenÃ­a


```python
# Diccionario con datos de una persona
persona = {
    "nombre": "MarÃ­a",
    "edad": 30,
    "ciudad": "Rosario"
}

# Usamos .get() para acceder de forma segura
print("Provincia:", persona.get("provincia"))  # No existe
print("Edad:", persona.get("edad"))

# Usamos .pop() para quitar una clave
ciudad_quitada = persona.pop("ciudad")
print("\nQuitamos la ciudad:", ciudad_quitada)

# Mostramos cÃ³mo quedÃ³ el diccionario
print("Diccionario actual:", persona)
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ  `.get("provincia")` devuelve `None` porque no existe esa clave, **sin lanzar error**.

ğŸ“Œ  `.pop("ciudad")` elimina esa clave y devuelve su valor original.

ğŸ’¡ `.get()` es ideal para **consultar sin riesgo**.  
ğŸ’¡ `.pop()` se usa para **sacar un dato que ya no necesitÃ¡s**.

ğŸ¯ ProbÃ¡ hacer `.get()` sobre otras claves inexistentes, o `.pop()` sobre claves que sÃ­ estÃ©n presentes.


# ğŸ“‡ Recorrer un diccionario con `.keys()`, `.values()` y `.items()`


Los mÃ©todos `.keys()`, `.values()` y `.items()` te permiten **ver y recorrer** el contenido de un diccionario:

- `.keys()` â†’ devuelve las claves
- `.values()` â†’ devuelve los valores
- `.items()` â†’ devuelve pares clave-valor


```python
# Diccionario con datos de una persona
persona = {
    "nombre": "MarÃ­a",
    "edad": 30,
    "email": "maria@gmail.com"
}

# Mostrar solo las claves
print("Claves:")
for clave in persona.keys():
    print(clave)

# Mostrar solo los valores
print("\nValores:")
for valor in persona.values():
    print(valor)

# Mostrar claves y valores juntos
print("\nClaves y valores:")
for clave, valor in persona.items():
    print(f"{clave}: {valor}")
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `.keys()` te deja recorrer solo los nombres de los datos.

ğŸ“Œ `.values()` te muestra los datos sin preocuparte por las claves.

ğŸ“Œ `.items()` es la forma mÃ¡s clara de recorrer un diccionario **clave por clave**.

ğŸ¯ ProbÃ¡ agregar mÃ¡s claves al diccionario y observÃ¡ cÃ³mo cambian los resultados.


# ğŸ“‡ `.setdefault(),` `.update()` y `.copy()`

Estos mÃ©todos te permiten manejar los diccionarios con mÃ¡s control:

- `.setdefault(clave, valor)` agrega un valor solo si **la clave no existe**
- `.update({clave: valor})` modifica o agrega varias claves a la vez
- `.copy()` crea una **copia independiente** del diccionario


```python
# Diccionario base
usuario = {
    "usuario": "techlab23",
    "nivel": "principiante"
}

# .setdefault(): solo agrega si la clave no existe
usuario.setdefault("email", "no_asignado@correo.com")
usuario.setdefault("nivel", "intermedio")  # No cambia nada

print("DespuÃ©s de setdefault:")
print(usuario)

# .update(): modifica o agrega valores
usuario.update({"nivel": "intermedio", "activo": True})

print("\nDespuÃ©s de update:")
print(usuario)

# .copy(): hacemos una copia independiente
copia_usuario = usuario.copy()
copia_usuario["usuario"] = "admin123"

print("\nCopia modificada:")
print(copia_usuario)

print("\nOriginal sin cambios:")
print(usuario)
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ `.setdefault()` **no sobreescribe**: solo crea si la clave no existe.

ğŸ“Œ `.update()` te permite modificar **muchas claves de una vez**.

ğŸ“Œ `.copy()` genera una nueva copia: los cambios sobre ella **no afectan al original**.

ğŸ¯ ProbÃ¡ eliminar `.copy()` y modificÃ¡ la copia directamente. Â¿Afecta al original?


# ğŸ“‡ Recorrer una lista de diccionarios

En muchas situaciones vas a tener **una lista de diccionarios**, cada uno con su propia informaciÃ³n.

Por ejemplo, una lista de personas, donde cada diccionario guarda los datos de una sola.


```python
# Lista de personas (cada una es un diccionario)
personas = [
    {"nombre": "Camila", "edad": 28},
    {"nombre": "DamiÃ¡n", "edad": 35},
    {"nombre": "SofÃ­a", "edad": 22}
]

# Recorremos la lista mostrando los datos
for persona in personas:
    print(f"{persona['nombre']} tiene {persona['edad']} aÃ±os.")
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ Cada elemento de la lista es un diccionario completo.

ğŸ“Œ Se accede a los datos internos con `persona["clave"]`, como siempre.

ğŸ“Œ Este formato es ideal para trabajar con registros de clientes, productos, fichas, etc.

ğŸ¯ ProbÃ¡ agregar una persona mÃ¡s o mostrar solo las mayores de 30 aÃ±os.


# DesafÃ­o -  Registro de productos con precios

TenÃ©s que construir un pequeÃ±o sistema para registrar productos que una persona va comprando.

Tu programa debe:

1. Usar un diccionario vacÃ­o al comenzar.
2. Pedir al usuario que cargue 3 productos con su precio.
3. Validar que el precio ingresado sea un nÃºmero mayor a 0.
4. Guardar el producto como **clave** y el precio como **valor**.
5. Mostrar el contenido completo del diccionario.
6. Mostrar el **total gastado**, sumando todos los precios.

ğŸ¯ Bonus: mostrar los productos en orden alfabÃ©tico (orden de claves).

(Intenta resolverlo antes de mirar la posible soluciÃ³n que aparece mÃ¡s abajo!)

```python
# Escribe aqui tu programa
```

----
Esta es una posible soluciÃ³n al desafÃ­o propuesto:

```python
# Diccionario vacÃ­o
productos = {}

# Cargamos 3 productos con su precio
for i in range(3):
    nombre = input(f"IngresÃ¡ el nombre del producto #{i + 1}: ").strip().lower()

    # ValidaciÃ³n del precio
    while True:
        try:
            precio = float(input(f"Precio de '{nombre}': $"))
            if precio > 0:
                break
            else:
                print("El precio debe ser mayor a 0.")
        except:
            print("IngresÃ¡ un nÃºmero vÃ¡lido.")

    productos[nombre] = precio  # Guardamos en el diccionario

# Mostramos los productos y precios
print("\nProductos cargados:")
for producto, precio in productos.items():
    print(f"{producto}: ${precio}")

# Calculamos el total
total = 0
for precio in productos.values():
    total += precio

print(f"\nTotal gastado: ${total}")

# Bonus: mostrar productos ordenados alfabÃ©ticamente
print("\nProductos en orden alfabÃ©tico:")
for producto in sorted(productos.keys()):
    print(producto)
```

# ğŸ” Â¿QuÃ© hace este cÃ³digo?

ğŸ“Œ Se usa un `for` para cargar tres productos y sus precios.

ğŸ“Œ El precio se valida: debe ser un nÃºmero y mayor a cero.

ğŸ“Œ Los productos se guardan en un diccionario: `clave = nombre`, `valor = precio`.

ğŸ“Œ `.items()` permite mostrar el contenido del diccionario.

ğŸ“Œ `.values()` se usa para acumular el total gastado.

ğŸ“Œ `sorted(productos.keys())` permite recorrer las claves ordenadas alfabÃ©ticamente.

ğŸ¯ Esta soluciÃ³n integra validaciÃ³n, acumulaciÃ³n, y ordenamiento de datos reales.
