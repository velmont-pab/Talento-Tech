Vamos a construir un programa que le pregunte a la persona su edad y le diga a qu√© grupo etario pertenece:

- Si tiene menos de 15 ‚Üí "Ni√±o/a"
- Si tiene entre 15 y 18 (inclusive) ‚Üí "Adolescente"
- Si tiene m√°s de 18 ‚Üí "Adulto/a"

Esto se puede hacer de forma clara usando `if`, `elif` y `else`.


```python
# Clasificaci√≥n por edad

edad = int(input("Ingres√° tu edad: "))

if edad < 15:
    print("Ni√±o/a")
elif edad <= 18:
    print("Adolescente")
else:
    print("Adulto/a")
```

üí° `elif` permite agregar m√°s de una condici√≥n intermedia.  
Python eval√∫a las condiciones de arriba hacia abajo y se detiene en la **primera verdadera**.

Ahora vamos a usar `match`, la estructura condicional que **simplifica el c√≥digo** cuando ten√©s que comparar un solo valor contra muchas opciones.

En este ejemplo, vamos a pedirle al usuario que ingrese una letra y vamos a decirle si es una **vocal**, una **consonante reconocida**, o si el valor es **inv√°lido**.


```python
# Clasificamos una letra usando match

letra = input("Ingres√° una letra: ").lower().strip()

match letra:
    case "a" | "e" | "i" | "o" | "u":
        print("Es una vocal.")
    case "b" | "c" | "d" | "f" | "g":
        print("Es una consonante com√∫n.")
    case _:
        print("No reconozco esa letra.")
```

üìå En este ejemplo:

- Usamos `.lower()` y `.strip()` para limpiar la entrada del usuario.
- Agrupamos m√∫ltiples casos en una sola l√≠nea con `|`, el operador ‚Äúo‚Äù de `match`.
- El caso `_` act√∫a como **‚Äúninguna de las anteriores‚Äù**.

üéØ Prob√° ingresar: "A", "f", "z", un n√∫mero o varios caracteres. ¬øQu√© ocurre?

# üîê Validando la longitud de una contrase√±a

En muchos sistemas se exige que la contrase√±a tenga una cantidad m√≠nima de caracteres para ser considerada segura.

En este bloque vamos a pedir una contrase√±a y validar si tiene al menos **8 caracteres**.


```python
# Solicitamos una contrase√±a
contrase√±a = input("Ingres√° tu contrase√±a: ")

# Validamos la longitud m√≠nima
if len(contrase√±a) < 8:
    print("La contrase√±a debe tener al menos 8 caracteres.")
else:
    print("Contrase√±a aceptada.")
```

üìå `len(contrase√±a)` devuelve la cantidad de caracteres que tiene la cadena ingresada.

üéØ Prob√° con contrase√±as cortas y largas, y observ√° c√≥mo reacciona el programa.

```python
# Pedimos una palabra
palabra = input("Ingres√° una palabra: ")

# Mostramos algunas partes de la cadena
print("Primera letra:", palabra[0])
print("√öltima letra:", palabra[-1])
print("Las tres primeras letras:", palabra[:3])
print("Las tres √∫ltimas letras:", palabra[-3:])
```

- Formatear el nombre y apellido para que se vean prolijos.
- Asegurarte de que el correo sea v√°lido (sin espacios y con un solo `@`).
- Clasificar a la persona por rango etario.
- Mostrar los datos correctamente con f-Strings.

----

Esta es una posible soluci√≥n al desaf√≠o propuesto:

```python
# Ingreso de datos
nombre = input("Nombre: ").strip().title()
apellido = input("Apellido: ").strip().title()
correo = input("Correo electr√≥nico: ").strip()
edad_texto = input("Edad: ").strip()

# Validaciones
correo_valido = "@" in correo and correo.count("@") == 1 and " " not in correo
edad_valida = edad_texto.isdigit()
mostrar_datos = correo_valido and edad_valida

if mostrar_datos:
    edad = int(edad_texto)

    # Clasificaci√≥n por edad
    if edad < 15:
        categoria = "Ni√±o/a"
    elif edad <= 18:
        categoria = "Adolescente"
    else:
        categoria = "Adulto/a"

    # Mostrar datos
    print("\n--- Ficha del cliente ---")
    print(f"Nombre: {nombre} {apellido}")
    print(f"Correo: {correo}")
    print(f"Categor√≠a: {categoria}")
else:
    print("ERROR: Los datos ingresados no son v√°lidos.")
```

üìå Este programa aplica todo lo aprendido:
- `.strip()` para limpiar espacios
- `.title()` para dar formato a nombres
- Validaci√≥n de correos con `count()` y `in`
- Clasificaci√≥n con `if / elif`
- Salida clara con **f-Strings**
